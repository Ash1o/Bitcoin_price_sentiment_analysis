```python
#columns which are not required
delete_cols = ['_type','url', 'id', 'quoteCount', 'conversationId', 'lang',
       'source', 'sourceUrl', 'sourceLabel', 'outlinks', 'tcooutlinks',
       'media', 'retweetedTweet', 'quotedTweet', 'inReplyToTweetId',
       'inReplyToUser', 'mentionedUsers', 'coordinates', 'place',
       'cashtags']

#deleting these columns 

tweet_df = data.drop(columns = delete_cols)
tweet_df.head()
```

splitting the datetime column to date and time column

```python
tweet_df['Dates'] = pd.to_datetime(tweet_df['date']).dt.date
tweet_df['Time'] = pd.to_datetime(tweet_df['date']).dt.time
tweet_df.head()
```

## now we need to apply various text processing techniques to clean the tweet data


### Preprocessing: Normalization
Words which look different due to casing or written another way but are the same in meaning need to be process correctly. Normalisation processes ensure that these words are treated equally. For example, changing numbers to their word equivalents or converting the casing of all the text.

```python
tweet_df['clean_t']=tweet_df['content'].str.lower()
```

```python
# removing all the special characters 
x = [ ]
for index, row in tweet_df.iterrows():
    sentence = row['clean_t']
    sentence = re.sub('[^a-z0-9$]',' ',sentence)
    x.append(sentence)

tweet_df['clean_t'] = x
```

```python
nltk.download('punkt')
```

```python
# stemming
sent2 = []
stemmer = PorterStemmer()
for index, row in tweet_df.iterrows():
    sent = row['clean_t']
    words = nltk.word_tokenize(sent)
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sent1 = ' '.join(words)
    sent2.append(sent1)
    
tweet_df['tweets_stem'] = sent2
```

```python
tweet_df[['clean_t','tweets_stem']].head(20)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer

sent4 = []
lemmatizer = WordNetLemmatizer()
for index, row in tweet_df.iterrows():
    sent = row['clean_t']
    words = nltk.word_tokenize(sent)
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sent3 = ' '.join(words)
    sent4.append(sent3)
    
tweet_df['tweets_lemm'] = sent4
```
