```python

import os
import pandas as pd
import json
```

```python
# Setting variables to be used in format string command below
tweet_count = 10000
text_query = "bitcoin min_faves:100 min_replies:25"
since_date = "2020-01-01"
until_date = "2021-01-01"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --since {} twitter-search "{} until:{}"> text-query-tweets1.json'.format(since_date, text_query, until_date))
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_df2 = pd.read_json('text-query-tweets1.json', lines=True)

# Displays first 5 entries from dataframe
tweets_df2
```

```python
tweets_df2.to_csv('bitcointweets_2020_2021.csv', sep=',', index=False)
```
